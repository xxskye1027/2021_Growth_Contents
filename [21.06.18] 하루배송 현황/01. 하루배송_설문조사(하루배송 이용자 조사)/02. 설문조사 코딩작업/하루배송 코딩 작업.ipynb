```python
df['코드 1'] = None
df['코드 2'] = None
df['코드 3'] = None
df['코드 4'] = None
```

```python
def re_turn(no, col):
    df.loc[df['No'] == no, col] = None
def no_show(no):
    return df.loc[df['No'] == no]
```

```python
def check(text):
    print(df.loc[df['하루배송_만족이유'].str.contains(text, na=False)].shape)
    return df.loc[df['하루배송_만족이유'].str.contains(text, na=False)]
def code(text, col, code):
    df.loc[df['하루배송_만족이유'].str.contains(text, na=False), col] = code
    print('done')
def no_code(no, col, code):
    df.loc[df['No'] == no, col] = code
```
