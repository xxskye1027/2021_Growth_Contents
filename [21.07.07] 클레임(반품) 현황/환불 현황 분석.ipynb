# 반품 유저 이탈율

- 환불 이후 이후 결제 완료일이 30일 이상 차이나는 유저

```python
del df['재구매소요일']
```

```python
df.insert(27, '재구매소요일', None)
```

```python
df.head()
```

```python
df['직후 결제완료일'].isnull().sum()
```

```python
for i in tqdm(df_list):
    df['재구매소요일'][i] = (df['최종 결제완료일'][i]-df['결제완료일'][i]).days
```

- 구매기반 이탈유저

- 직후 상품구매가 최종 결제완료랑 동일한 경우는 있어도, 직후가 반품한 상품 결제완료 일자보다 이른 일자면 안되는데..?
- 널값이 있어서 마이너스까지 잡은거지만, 일자가 이른 일자면 해당 열이 정확하지 않다는 것..
- 직후 결제완료일 정합성 낮음.. 최종 결제완료일로 작업 진행

```python
- 마이너스의 경우 완전 이탈
- 양수일 경우 재구매는 진행
- 양수인 고객 중 최종 결제 완료일에서 어느 정도 소요됐는지
```
